```python
def get_square(arr,r,c):
    
    width=70
    height=70
    
    left_side=45
    top_side=30
    
    square=arr[(top_side+height*c):(top_side+height*(c+1)),(left_side+width*r):(left_side+width*(r+1)),:]
    return square
                                                                               
                                                                               
```

```python
imshow(get_square(arr,1,1))
```
