```python
from bbturtle import *
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)
    
```

```python
square(10)
```

```python
reset((8,8))

penup()
goto(-75,0)
pendown()

color='red'
for i in range(10):
    pencolor(color)
    square(10)

    penup()
    forward(15)
    pendown()

    if color=='red':
        color='black'
    else:
        color='red'
    

```
